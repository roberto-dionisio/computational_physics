```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
%config InlineBackend.figure_format= 'retina'
```

```python
def f(x):
    return 3.0*x**3+x-3
```

```python
def g(x):
    #return f(x)+x
    return ((3.0-x)/3.0)**(1/3)
```

```python
def dg(x):
    #return 9*x**2+2
    return -1.0/9.0*(3.0/(3.0-x))**(2/3)
```

```python
def fixpt(x0,es,imax):
    xr=x0
    x=[];e=[];it=[]
    for i in range(imax):
        xrold=xr
        xr=g(xrold)
        if(xr!=0):
            ea=abs((xr-xrold)/xr)*100
        it.append(i);e.append(ea); x.append(xr)
        if(ea<es):
            break
    return it,x,e
```

```python
i3,x3,e3=fixpt(2,1e-12,100)
```

```python
plt.plot(i3,e3,label='Fixpnt method')
plt.yscale('log')
plt.grid()
plt.show()
print(x3)
```
