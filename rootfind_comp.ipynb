```python
import numpy as np
%matplotlib inline
```

```python
def f(x):
    return 3.0*x**3+x-3
```

```python
def g(x):
    #return f(x)+x
    return ((3.0-x)/3.0)**(1/3)
```

```python
def dg(x):
    #return 9*x**2+2
    return -1.0/9.0*(3.0/(3.0-x))**(2/3)
```

```python
def df(x):
    return 9*x**2+1
```

```python
def bisect(xl,xu,es,imax):
    xr=0.0
    x=[];e=[];it=[]
    for i in range(imax):
        xrold=xr
        xr=0.5*(xl+xu)
        if(xr!=0.0):
            ea=abs((xr-xrold)/xr)*100
        it.append(i);e.append(ea);x.append(xr)
        test=f(xl)*f(xr)
        if (test<0.00):
            xu=xr
        elif(test>0.0):
            xl=xr
        else:
            ea=0.0
        if(ea<es):
            break
    return it,x,e
```

```python
i1,x1,e1=bisect(0.2,2.453,1e-12,100000000)
plt.plot(i1,x1,label="Bisection")
plt.yscale('linear')
plt.grid()
plt.show()
```

```python
def falsep(xl,xu,es,imax):
    xr=0.0
    x=[];e=[];it=[]
    for i in range(imax):
        xrold=xr
        xr=xu-f(xu)*(xl-xu)/(f(xl)-f(xu))
        if (xr!=0.0):
            ea=abs((xr-xrold)/xr)*100
        it.append(i);e.append(ea);x.append(xr)
        test=f(xl)*f(xr)
        if (test>0.0):
            xl=xr
        elif (test<0.0):
            xu=xr
        else:
            ea=0.0
        if (ea<es):
            break
    return it,x,e         
```

```python
i2,x2,e2=falsep(0.2,2.453,1e-12,1000)
plt.plot(i1,x1,label="Bisection")
plt.yscale('linear')
plt.grid()
plt.show()
```

```python
def nr(x0,es,imax):
    x=[];it=[];e=[]
    xr=x0
    
    for i in range(imax):
        xrold=xr
        xr=xrold-f(xrold)/df(xrold)
        if (xr!=0):
            ea=abs((xr-xrold)/xr)*100
        x.append(xr);it.append(i);e.append(ea)
        if (ea<es):
            break
    return it,x,e
```

```python
i3,x3,e3=nr(10,1e-12,1000)
plt.plot(i3,e3,label="Nr")
plt.plot(i1,e1,label='Bsc')
plt.yscale('log')
plt.legend()
plt.grid()
plt.show()
```
